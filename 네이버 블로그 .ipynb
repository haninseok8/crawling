```python
import re
import json
import requests
import urllib.request
import urllib.error
import urllib.parse
from bs4 import BeautifulSoup
import pandas as pd

naver_client_id = "nm5K2BIGoKdpTlQXB00D"
naver_client_secret = "1TiclHife0"

def blog_link(keyword,num):

    df_blog = pd.DataFrame()
    keyword =  urllib.parse.quote(keyword)
    
    for i in range(num):
        start = i*100+1
        url = "https://openapi.naver.com/v1/search/blog?query=" + keyword +"&display=100" +"&start="+ str(start) #json 결과
        # url = ”https://openapi.naver.com/v1/search/blog.xml?query=" + encText # xml 결과

        request = urllib.request.Request(url)
        request.add_header("X-Naver-Client-Id",naver_client_id)
        request.add_header("X-Naver-Client-Secret",naver_client_secret)
        response = urllib.request.urlopen(request)
        rescode = response.getcode()

        if(rescode==200):
            response_body = response.read()
            json_data = json.loads(response_body.decode('utf-8'))
            df = pd.DataFrame(json_data['items'])
            df_link = df[['title','postdate','link']]
            if df_blog.empty :
                df_blog = df_link
            else:
                df_blog = pd.concat([df_blog, df_link],ignore_index=False)
               # print(response_body.decode('utf-8'))
        else:
            print("Error Code:" + rescode)
            
    return df_blog


```

```python
keyword = input("검색할 단어 : ")
num = input("갯수(약 *100) : " )
df_blog = blog_link(keyword,int(num))
```

```python
from urllib.request import urlretrieve
from urllib.parse import quote_plus    # korean support
from bs4 import BeautifulSoup as bs    # Essential module
from selenium import webdriver         # Google crolling
import time

driver= webdriver.Chrome('chromedriver.exe')
options = webdriver.ChromeOptions()
options.add_experimental_option('excludeSwitches', ['enable-logging'])
df_blog['content'] = ""

for i in range(len(df_blog)):
    
    texts=""
    url = df_blog.iloc[i]['link']           
    driver.get(url)
    try:
        driver.switch_to.frame('mainFrame')

        currentURL = driver.current_url
        idnum = currentURL.split('/')[-1]

        html = driver.page_source
        soup = bs(html,features="html.parser")

        div = soup.find("div",id="post-view"+idnum)
        p = div.find_all("p")

        for text in p:
            texts = texts+text.text

        df_blog['content'][i] = texts
    
    except:
        print(str(i)+"번째 블로그 오류")
        
    
print("------Done------")
        
```
